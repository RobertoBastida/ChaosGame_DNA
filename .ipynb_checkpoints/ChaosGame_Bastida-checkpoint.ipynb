### Definición de las variables

Para la iteración sobre la longitud de la sub-cadena de caracteres.

```python
power = [10000,100000,1000000,10000000,100000000]
dic_dim = {}
```

## Cromosoma 19

```python
Chr = "chr_19.txt"
Str_Chr = load_txt_file("data/",Chr)
chr_dimension = []
for l in power:
    # trimming the string to a length of maximum 1,000,000 nucleotides
    CG_Chr = chaos_game_chromosome(Str_Chr[0:min(len(Str_Chr),l)],Dic_2)

    # grabbing the coordinates
    CG_x = [i[0] for i in CG_Chr]
    CG_y = [i[1] for i in CG_Chr]

    # Plotting the series
    if l == power[0]:
        plt.scatter(CG_x,CG_y,s=0.0125)
    elif l == power[1]:
        plt.scatter(CG_x,CG_y,s=0.000125)
    elif l == power[-1]:
        plt.scatter(CG_x,CG_y,s=0.0000095)
    else:
        plt.scatter(CG_x,CG_y,s=0.0000125)
        
    plt.axis('equal')
    plt.axis('off')
    plt.savefig("figurasprueba/plot/ch19/{}".format(Chr.split(".")[0]+"_"+str(power.index(l))),dpi=200)
    plt.show()
    plt.clf()

    Chr_Fd = Frac_dimension_computation_name(CG_x,CG_y,"figurasprueba/dim/ch19/{}".format(Chr.split(".")[0]+"_"+str(power.index(l))))
    chr_dimension.append(Chr_Fd[-1])
dic_dim[Chr] = {i:j for i,j in zip(power,chr_dimension)}
```

## Emulando distribuciones de nucleótidos

```python
rand_chr_dimension = []
for l in power:
    Str_Chr = random_nuc(l)
    
    # trimming the string to a length of maximum nucleotides
    CG_Chr = chaos_game_chromosome(Str_Chr[0:min(len(Str_Chr),l)],Dic_2)

    # grabbing the coordinates
    CG_x = [i[0] for i in CG_Chr]
    CG_y = [i[1] for i in CG_Chr]

    # Plotting the series
    if l == power[0]:
        plt.scatter(CG_x,CG_y,s=0.0125)
    elif l == power[1]:
        plt.scatter(CG_x,CG_y,s=0.000125)
    elif l == power[-1]:
        plt.scatter(CG_x,CG_y,s=0.00000125)
    else:
        plt.scatter(CG_x,CG_y,s=0.0000125)
        
    plt.axis('equal')
    plt.axis('off')
    plt.savefig("figurasprueba/plot/random/random_{}".format(power.index(l)),dpi=200)
    plt.show()
    plt.clf()

    # Computing the Minkowski-Bouligand dimension
    Chr_20_Fd = Frac_dimension_computation(CG_x,CG_y,"figurasprueba/dim/random/{}".format(power.index(l)),dpi=150)
    rand_chr_dimension.append(Chr_20_Fd[-1])
    
# Appeding dimension to the dimension dictionary
dic_dim["Random_chr"] = {i:j for i,j in zip(power,rand_chr_dimension)}
```
